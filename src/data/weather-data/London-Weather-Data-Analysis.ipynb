```python
#Import libraries
import pandas as pd
import datetime
import numpy as np
```

```python
#load data for the different years
data_2019 = pd.read_csv("C:/Users/HP/Downloads/London_2019_with_daylight_hours.csv")
data_2020 = pd.read_csv("C:/Users/HP/Downloads/London_2020_with_daylight_hours.csv")
data_2021 = pd.read_csv("C:/Users/HP/Downloads/London_2021_with_daylight_hours.csv")
```

```python
#Concatenate data to have one data set
data = pd.concat([data_2019, data_2020, data_2021])
data.head()
```

```python
#Check the data types of columns
data.info()
```

```python
#Create a date column
data["DATE"] = data.apply(lambda x: datetime.date(x['Year'], x['Month'], x['Date']), axis=1)
data.head()
```

```python
#Drop unneccessary columns
data.drop(columns=["Date", "Month", "Year", "Time"], inplace = True)
data.head()
```

```python
data.columns
```

```python
#Change order of columns
column_order = ["DATE", 'Low (in Celsius)', 'High (in Celcius)', 'weather_desc', 'daylight_hours', 
                'daylight_mins', 'daylight_sec']

data = data.reindex(columns=column_order)
data.head()
```

```python
#Group data by date

#get the description for each day
dates = data.groupby("DATE")

description = []
for date in dates:
    description.append(date[1]['weather_desc'].mode()[0])
    #print(date[0], date[1]['description'].mode()[0])
    
description
```

```python
#Find the daily temperature averages
data = data.groupby("DATE").mean().reset_index()

#add description column
data['weather_desc'] = description
data['Average Temperature'] = data.apply(lambda x: (x['Low (in Celsius)'] + x['High (in Celcius)'])/2, axis=1)
data.rename(columns={'DATE':'Date'}, inplace=True)

data.head()
```

```python
data.columns
```

```python
#Change order of columns
column_order = ["Date", 'Low (in Celsius)', 'High (in Celcius)', 'Average Temperature', 'weather_desc', 
                'daylight_hours', 'daylight_mins', 'daylight_sec']

data = data.reindex(columns=column_order)
data.head()
```

```python
data
```

```python
#set index as the "Date" column
data.set_index("Date", inplace=True)
data.index = pd.DatetimeIndex(data.index)
```

```python
#Check for any missing values
data.isnull().sum()

#There is none, therefore we can go on.
```

```python
data.head()
```

```python
#Save the data as a csv file
data.to_csv("C:/Users/HP/Desktop/Omdena/Mental Health Project/London_Daily_Weather_Data.csv", 
            index=True, date_format='%Y-%m-%d')
```

#### Monthly Averages

1. __Temperature__

```python
#Monthly temperature averages
temp = data['Average Temperature'].resample("M").mean()
temp = pd.DataFrame(round(temp, 2))
temp.head()
```

```python
#Check if there are missing values
temp["Average Temperature"].isnull().sum()

#Get the observation with missing value
temp[temp["Average Temperature"].isnull()]
```

```python
#Fill the missing value using the mean of the whole year 2020
temp.fillna(temp.iloc[12:24, ].mean(), inplace=True)
```

```python
#Check if there is still any missing values
temp["Average Temperature"].isnull().sum()

#We're good to go
```

```python
#Save the data as a csv file
temp.to_csv("C:/Users/HP/Desktop/Omdena/Mental Health Project/London_Monthly_Temperature_Averages.csv", 
            index=True, date_format='%Y-%m-%d')
```
