```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('quarterlyalcoholspecificdeaths.xlsx', sheet_name = 'Table 4', header=[23,24])
df.head()
```

```python
df = df.dropna(axis=1, how='all')
df.head()
```

```python
df = df.dropna(axis=1)
df.head()
```

```python
df.columns = df.columns[:4].droplevel(level=0).append(df.columns[4:].map('_'.join))
```

```python
df
```

```python
df2 = df[df['Year of death registration'] == 2020]
df2.head()
```

```python
df3 = df2[df2['Sex'] == 'Persons']
df3
```

```python
df3 = pd.wide_to_long(df2, 
                stubnames = ['Q1','Q2','Q3','Q4'], 
                i=['Sex','Area name'], 
                j='Indicator', 
                suffix= '.+', 
                sep='_')
df3
```

```python
df4 = df3.reset_index().drop(['Year of death registration','Area code'], axis=1)
df4
```

```python
df5 = df4.melt(id_vars=['Sex','Area name','Indicator'], value_vars=['Q1','Q2','Q3','Q4'], var_name='Quarter')
df5
```

```python
df6 = df5[df5['Indicator'] == 'Deaths']
df6 = df6.drop('Indicator', axis=1)
df6
```

```python
df7 = df6.pivot_table(index=['Quarter','Sex'],columns='Area name',values='value', aggfunc='sum')
df7.columns.name = None
df7
```

```python
df8 = df7.reset_index()
df8
```

```python
df9 = df6.pivot_table(index=['Quarter'],columns=['Area name','Sex'],values='value', aggfunc='sum')
df9.columns.name = None
df9
```

```python
df10 = df6[df6['Sex'] == 'Persons']
df10
```

```python
df11 = df10.pivot_table(index = 'Quarter', columns='Area name', values='value')
df11.columns.name = None
df11 = df11.reset_index()
df11
```

```python
df12 = df11.reindex(columns = ['Quarter', 'London', 'North East', 'North West','East Midlands','West Midlands',
                      'East of England', 'South East', 'South West', 'Yorkshire and The Humber','England', 'Wales'])
```

```python
df12.to_csv('final_quarterlyregional.csv', index=False)
```
