## **Defining a Filter Function**

```python
def NHS_filter(path, column='PRIMARY_LEVEL_DESCRIPTION', location= 'london'):
  """Filters the location London form the dataset within a specified column.
  INPUT:
  path: str. The path from local or remote directory.
  column: str.The specified column header in the dataset.
  location: str. The required location contained in each data point within the colomn specified.
  """
  filt = df[column].str.lower()
  filt = filt.str.contains(location, na = False)
  data = df.loc[filt]
  return data
```

```python
new_df = NHS_filter(path)
new_df.head()
```

```python
# Comparing the difference between the total dataset and that for London.
print('The number of observations for London: {}'.format(len(new_df)))
print('-'*50)
print('The number of observation for England: {}'.format(len(df)))
```
